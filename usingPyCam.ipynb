```python
import subprocess
import cv2
import matplotlib.pyplot as plt
import time
```

```python
def capture():
    result = subprocess.call("./capture.sh")
```

```python
def FrameCapture():
    fig = plt.figure(figsize=(20, 10))
    obj = cv2.CascadeClassifier('/home/pi/Desktop/DL/35.xml')
    
    path = "/home/pi/Desktop/DL/captures/lol.jpg"
    
    image = cv2.imread(path)
    print(image.shape)
    img = image.copy()
    img = cv2.resize(img, (0, 0), fx = 0.1, fy = 0.1)
    img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    res = obj.detectMultiScale(img2,minNeighbors=5,minSize=[40,40])
        
    if len(res) == 0:
        print("no bus found")
        fig.add_subplot(1, 2, 1)
        im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(im_rgb)
    else:
        for (x,y,w,h) in res:
            print("found Bus")
            start, end = (x, y), (x+w, y+h)
            cv2.rectangle(img, pt1=start, pt2=end, color=(255,255,0), thickness=2)
            fig.add_subplot(1, 2, 1)
            im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            plt.imshow(im_rgb)
```

```python
start = time.time()

capture()

mid = time.time()

FrameCapture()

end = time.time()
print("time taken to capture photo : ", mid-start)
print("time taken to process photo : ", end-mid)
```
