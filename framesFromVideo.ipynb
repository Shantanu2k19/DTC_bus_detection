```python
obj = cv2.CascadeClassifier('/home/zodiac/Desktop/3.xml')


def FrameCapture(path):
    vidObj = cv2.VideoCapture(path)

    count = 0
    z=0
    fig = plt.figure(figsize=(20, 100))

    success = 1

    while success:
        success, image = vidObj.read()
        if(count%15==0):
            z+=1
            img = image
            img = img[420:1500, 0:1080]
            img = cv2.resize(img, (0, 0), fx = 0.1, fy = 0.1)
            print(img.shape)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            res = obj.detectMultiScale(img)
            
            if len(res) == 0:
                fig.add_subplot(20, 5, z)
                img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                plt.imshow(img2)
            else:
                for (x,y,w,h) in res:
                    img2 = img.copy()
                    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
                    start, end = (x, y), (x+w, y+h)
                    cv2.rectangle(img2, pt1=start, pt2=end, color=(255,255,0), thickness=2)
                    fig.add_subplot(20, 5, z)
                    plt.imshow(img2)
            return
        if(z>80):
            return
        count += 1


name = "testVid1.mp4"        
direc = "/home/zodiac/Desktop/rtai/notebooks/data2"
test_vid = direc+name
FrameCapture(name)
```
